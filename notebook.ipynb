```python
from simulater import Simulater
simulater = Simulater(
    time_step=1,
    target_temperature=200.0,
    initial_temperature=200.0,
    tolerance=10.0,
    conveyor_length=1000,
    newton_cooling_coefficient=0.01,
    natural_temperature=20.0,
    mean_conveyor_speed=2.0
)
```

```python
heating_point = [100,200,300,500,700]
heating_power_coef = [15,20,10,15,10]

# heating_point = [20,60,100,140,180,220,260,300,340,380,420,460,500,540,580,620,660,700,740,780,820,860,900,940,980]
# heating_power_coef = [9] * 25

# heating_point = []
# heating_power_coef = []

simulater.draw_plot_with_simulator(
    heating_point=heating_point, 
    heating_power_coef=heating_power_coef
)
```

```python
df, actual_temperatures, product_positions, times, speed_history = simulater.simulate(
    heating_point=heating_point, 
    heating_power_coef=heating_power_coef
)
```

```python
def total_loss(actual_temperatures: list[float],
            target_temperature: float,
            times: int) -> float:
    return sum([abs(target_temperature - temperature) for temperature in actual_temperatures]) / times

total_loss(actual_temperatures, 200.0, len(times))
```
